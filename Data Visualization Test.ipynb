```python
dfCon2 = dfCon2.drop(dfCon2[dfCon2.LifeExpectancy.isna() == True].index, axis=0)
```

```python
dfCon2.describe()
```

```python
dfCon2.LifeExpectancy.describe()
```

```python
Q1 = dfCon2.describe()['LifeExpectancy']['25%']
Q2 = dfCon2.describe()['LifeExpectancy']['50%']
Q3 = dfCon2.describe()['LifeExpectancy']['75%']
IQR = Q3-Q1

meanAsia = dfCon2.describe()['LifeExpectancy']['mean']

minBound = Q1 - (1.5 * IQR)
maxBound = Q3 + (1.5 * IQR)

stdAsia = dfCon2.describe()['LifeExpectancy']['std']

print(f'IQR LifeExpectancy Asia : {IQR}')
print(f'Batas minimum : {minBound}')
print(f'Batas maksimum : {maxBound}')
print(f'Mean : {meanAsia}')
print(f'Median : {Q2}')
print(f'Standard Deviation : {stdAsia}')
```

```python
sns.distplot(dfCon2['LifeExpectancy'])

plt.axvline(dfCon2.describe()['LifeExpectancy']['mean'], color='green')
plt.axvline(dfCon2.describe()['LifeExpectancy']['50%'], color='red')
plt.axvline(dfCon2.describe()['LifeExpectancy']['std'], color='yellow')
plt.axvline(minBound, color='purple')
plt.axvline(maxBound, color='purple')


```
